# Question-2(a)

```python
import numpy as np
array = np.array([10,52,62,16,16,54,453])
sorted_array = np.sort(array)
print(f"sorted array = {sorted_array}")

sorted_indices = np.argsort(array)
print(f"indices of srted array = {sorted_indices}")

print("4 smallest element in array =")
for i in range(0,4):
  print(sorted_array[i],end=" ")

rev_sorted_array = np.sort(array)[::-1]
print(f"sorted array = {rev_sorted_array}")
print("5 largest element in array")
for i in range(0,5):
  print(rev_sorted_array[i],end=" ")
```

# Question-3(c)

```python
import numpy as np

# Given
X = 131
sales = np.array([131, 181, 231, 281, 331])

# Apply discount based on condition using broadcasting
discounted_sales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)

print("Discounted Sales:\n", discounted_sales)
```

# Question-3(d)

```python
# Stack the discounted sales 3 times to simulate 3 weeks
weekly_sales = np.stack([discounted_sales] * 3)

# Increase sales by 2% per week
increase_factors = np.array([1.00, 1.02, 1.04]).reshape(3, 1)  # shape (3,1) for broadcasting
expanded_sales = weekly_sales * increase_factors

print("Expanded Weekly Sales:\n", expanded_sales)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(-10,10,100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(abs(x)+1)

plt.plot(x,y1,label = "y=x^2",linestyle = "-")
plt.plot(x,y2,label = "y=sin(x)",linestyle = "--")
plt.plot(x,y3,label = "y=exp^x",linestyle = "-.")
plt.plot(x,y4,label = "y=log(|x|+1)",linestyle = ":")

plt.ylim(0,10)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plots of Different Functions")
plt.legend()
plt.grid(True)
plt.show()
```
