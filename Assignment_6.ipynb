```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
M = int(input("enter any value of M:"))
y1 = M*x**2
y2 = M*np.sin(x)

plt.plot(x,y1,label = "y = M(x^2)",linestyle = "--",color = "red")
plt.plot(x,y2,label = "y = M(sin(x))",linestyle = ":",color = "green")
plt.legend()
plt.grid(True)
plt.show()
```

# Question-2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset
data = {
    'Subjects': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Scores': [88, 75, 92, 85, 95]
}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 5))
barplot = sns.barplot(x='Subjects', y='Scores', data=df, palette='muted')

# Add annotations
for i, score in enumerate(df['Scores']):
    barplot.text(i, score + 1, str(score), ha='center')

plt.title("Student Scores in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True, axis='y')
plt.show()
```

# Question-3

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace with your roll number
roll_number = 123456
np.random.seed(roll_number)

data = np.random.randn(50)

# Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot - cumulative sum
axs[0, 0].plot(np.cumsum(data), color='purple')
axs[0, 0].set_title("Cumulative Sum")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumsum")
axs[0, 0].grid(True)

# Scatter plot - random noise
axs[0, 1].scatter(range(50), data, color='green')
axs[0, 1].set_title("Random Noise")
axs[0, 1].set_xlabel("Index")
axs[0, 1].set_ylabel("Value")
axs[0, 1].grid(True)

# Leave bottom 2 subplots empty
axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

# 1. Total Profit Line Plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

# 2. Multi-line Product Sales Plot
plt.figure(figsize=(10, 6))
for column in df.columns[1:-1]:  # skip 'month_number' and 'total_profit'
    plt.plot(df['month_number'], df[column], label=column)
plt.title("Product Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all features
df_sum = df.sum(numeric_only=True)[1:]  # sum of all attributes (excluding month)
plt.figure(figsize=(10, 5))
sns.barplot(x=df_sum.index, y=df_sum.values, palette='viridis')
plt.title("Total Sales by Product")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.ylabel("Total Sales")
plt.show()

```
